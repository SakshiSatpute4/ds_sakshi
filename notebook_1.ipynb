```python
# --- Step 1: Import libraries ---
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.model_selection import train_test_split

# Make graphs more specific
sns.set(style="whitegrid")
plt.rcParams['figure.figsize'] = (8,5)

# --- Step 2: Load datasets (already uploaded in Colab) ---
hist = pd.read_csv("historical_data.csv")
fg = pd.read_csv("fear_greed_index.csv")

print("Historical data shape:", hist.shape)
print("Fear/Greed data shape:", fg.shape)

# --- Step 3: Clean column names (remove spaces, lowercase) ---
hist.columns = hist.columns.str.strip().str.lower().str.replace(" ", "_")
fg.columns = fg.columns.str.strip().str.lower().str.replace(" ", "_")

print("\nCleaned historical columns:", hist.columns.tolist())
print("Cleaned fear/greed columns:", fg.columns.tolist())

# --- Step 4: Parse datetime columns ---
hist['timestamp_ist'] = pd.to_datetime(hist['timestamp_ist'], errors='coerce', format="%d-%m-%Y %H:%M")
hist['date_only'] = hist['timestamp_ist'].dt.date

fg['date'] = pd.to_datetime(fg['date'], errors='coerce')
fg['date_only'] = fg['date'].dt.date

# --- Step 5: Merge datasets on date ---
merged = pd.merge(
    hist,
    fg[['date_only','classification']],
    on='date_only',
    how='left'
)

print("\nMerged dataset shape:", merged.shape)
display(merged.head())

# --- Step 6: Feature Engineering ---
merged['side_enc'] = merged['side'].map({'BUY':1, 'SELL':-1})
merged['closed_pnl'] = pd.to_numeric(merged['closed_pnl'], errors='coerce')
merged['size_usd'] = pd.to_numeric(merged['size_usd'], errors='coerce')

# --- Step 7: Aggregated view ---
agg = merged.groupby(['account','classification']).agg({
    'closed_pnl':'mean',
    'size_usd':'sum',
    'execution_price':'mean',
    'side_enc':'mean'
}).reset_index()

print("\nAggregated trader behavior by sentiment:")
display(agg.head())

# --- Step 8: Visualizations ---

# 1. PnL distribution
plt.figure(figsize=(8,5))
sns.boxplot(x='classification', y='closed_pnl', data=merged)
plt.title("Trader Profit/Loss vs Market Sentiment")
plt.show()

# 2. Trade Volume
plt.figure(figsize=(8,5))
sns.barplot(x='classification', y='size_usd', data=merged, estimator=np.mean, ci=None)
plt.title("Average Trade Volume vs Sentiment")
plt.show()

# 3. Side distribution
plt.figure(figsize=(8,5))
sns.countplot(x='classification', hue='side', data=merged)
plt.title("Trade Side (Buy/Sell) vs Sentiment")
plt.show()

# --- Step 9: Simple predictive model ---
model_df = merged.dropna(subset=['classification','closed_pnl','size_usd','side_enc'])
le = LabelEncoder()
model_df['sentiment_enc'] = le.fit_transform(model_df['classification'])

X = model_df[['closed_pnl','size_usd','execution_price','side_enc']]
y = model_df['sentiment_enc']

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=42)

model = LogisticRegression(max_iter=500)
model.fit(X_train,y_train)

y_pred = model.predict(X_test)

print("\nClassification Report:\n", classification_report(y_test,y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test,y_pred))
```
